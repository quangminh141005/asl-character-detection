```python
from pathlib import Path

root = Path("/content/hand-dataset-v2/Hand-dataset-v2")  # change this to your dataset root
yaml_path = root / "hand-dataset-v2.yaml"

yaml_text = f"""# YOLOv8 data config for hand detection
path: {root}

train: images/train
val: images/val
test: images/test

names:
  0: hand
"""

with open(yaml_path, "w") as f:
    f.write(yaml_text)

print("YAML saved to:", yaml_path)
```

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")

model.train(
    data="/content/hand-dataset-v2/Hand-dataset-v2/hand-dataset-v2.yaml",
    epochs=50,
    imgsz=512,
    batch=-1,
    device=0,
    workers=8,
    single_cls=True
)
```
